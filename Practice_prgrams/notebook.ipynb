```python
class soluition:
    def twosum(self, nums:list[int], target:int) -> list[int]:
        new_num = nums.copy()
        new_num.sort()
        print(f"sorted values {new_num}")

        for i in range(len(new_num)):
            second_num = target - new_num[i]
            if second_num in new_num and new_num.index(second_num) !=i:
                #print(f"first num {new_num[i]} and second num {second_num}")
                return [nums.index(new_num[i]), nums.index(second_num)]

s1 = soluition()
s1.twosum([5, 3, 7 ,8, 2, 4, 3], 6)
```

```python
class solution:
    def sell_or_buy(self, prices:list[int]) -> int:
        #[3,7,5,2,1,3,6,4]
        min_price = prices[0]
        max_profit =0
        profit_sum =0
        
        for i in range(1, len(prices)):
            if prices[i] > min_price:
                if max_profit < (prices[i] - min_price):
                    max_profit = prices[i] - min_price
                    profit_sum += max_profit
                print(f"min price {min_price} and current price {prices[i]} and max profit {max_profit} ")
            else:
                min_price = prices[i]
                print(f"min price {min_price} and current price {prices[i]} and max profit {max_profit} ")
        # return max_profit,profit_sum
        return max_profit, profit_sum        

s2 = solution()
print(s2.sell_or_buy([3,7,5,2,1,3,6,4]))
```

```python
class soluition:
    def sell_buy(self, prices):
        max_profit = 0
        #3,7,9,5,2,1,3,6,4
        #7-3 = 4
        #9-7 = 2
        #3-1 = 2
        #6-3 = 3
        for i in range(1,len(prices)):
            if prices[i] > prices[i-1]:
                max_profit =max_profit + (prices[i] - prices[i-1])
        return max_profit
s3 = soluition()
print(s3.sell_buy([3,7,9,5,2,1,3,6,4]))
```

```python
## solution to find the unique value in th elist
class soluition():
    def find_uniq(self, arr):
        result=0
        for i in arr:
            result = result ^ i
        return result

s4 = soluition()
print(s4.find_uniq([1,1,1,2,2,2,3,3,3,4]))
```

```python
## count the majority element in an array
class solution:
    def majority_element(self, nums:list[int]) -> int:
        count = 0
        majority = 0

        for num in nums:
            if count == 0:
                majority = num
            count += (1 if num == majority else -1)
        return majority

s5 = solution()
print(s5.majority_element([2,2,1,1,1,2,2]))
```

```python
# remove duplicated from sorted array
#0,0,1,1,1,2,2,3,3,4]
class solution:
    def remove_dup(self, nums):

        size =len(nums)
        if size == 0:
            return 0
        insertindex = 1
        for i in range(1, size):
            if nums[i] != nums[i-1]:
                nums[insertindex] = nums[i]
                insertindex +=1
        return insertindex, nums
s6 = solution()
print(s6.remove_dup([0,0,1,1,1,2,2,3,3,4]))
```

```python
# rotate an array to the right by k steps
class soluition:
    def rotate_arr(self, nums,k):
        k = k % len(nums)
        nums[:] = nums[-k:] + nums[:-k]
        return nums
s7 = soluition()
print(s7.rotate_arr([1,2,3,4,5,6,7], 3))
```
