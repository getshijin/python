```python
class soluition:
    def twosum(self, nums:list[int], target:int) -> list[int]:
        new_num = nums.copy()
        new_num.sort()
        print(f"sorted values {new_num}")

        for i in range(len(new_num)):
            second_num = target - new_num[i]
            if second_num in new_num and new_num.index(second_num) !=i:
                #print(f"first num {new_num[i]} and second num {second_num}")
                return [nums.index(new_num[i]), nums.index(second_num, nums.index(second_num) + 1)]

s1 = soluition()
s1.twosum([5, 3, 7 ,8, 8, 4, 3], 6)
```

```python
class solution4:
    def sumoftwo(self,nums,target):
        num_copy= nums.copy()
        num_copy.sort()
        print(f"sorted values {num_copy}")
        i=0
        j=len(num_copy)-1
        for i in range(len(num_copy)):
            while i < j:
                #print(f"first num {num_copy[i]} and second num {num_copy[j]}")
                sum = num_copy[i] + num_copy[j]
                if sum == target:
                    return [nums.index(num_copy[i]), nums.index(num_copy[j], nums.index(num_copy[i])+1)]
                elif sum < target:
                    i +=1
                else:
                    j -=1

        return 1
s2 = solution4()
result =s2.sumoftwo([5, 3, 7 ,8, 4, 4, 3], 6)
print(result)
```

```python
class solution:
    def sell_or_buy(self, prices:list[int]) -> int:
        #[10,3,7,9,5,2,1,3,6,4]
        min_price = prices[0]
        max_profit =0
        
        
        for i in range(1, len(prices)):
            if prices[i] >= prices[i-1]:
                continue
            else:
                
                max_profit = max_profit + (prices[i-1] - min_price)
                min_price = prices[i]

        return max_profit        

s2 = solution()
print(s2.sell_or_buy([10,3,7,9,5,2,1,3,6,4]))
```

```python
## solution to find the unique value in th elist
class soluition():
    def find_uniq(self, arr):
        result=0
        for i in arr:
            result = result ^ i
        return result

s4 = soluition()
print(s4.find_uniq([1,1,1,2,2,2,3,3,3,4]))
```

```python
## count the majority element in an array
class solution:
    def majority_element(self, nums:list[int]) -> int:
        count = 0
        majority = 0

        for num in nums:
            if count == 0:
                majority = num
            count += (1 if num == majority else -1)
        return majority

s5 = solution()
print(s5.majority_element([2,2,1,1,1,2,2]))
```

```python
# remove duplicated from sorted array
#0,0,1,1,1,2,2,3,3,4]
class solution:
    def remove_dup(self, nums):

        size =len(nums)
        if size == 0:
            return 0
        insertindex = 1
        for i in range(1, size):
            if nums[i] != nums[i-1]:
                nums[insertindex] = nums[i]
                insertindex +=1
        return insertindex, nums
s6 = solution()
print(s6.remove_dup([0,0,1,1,1,2,2,3,3,4]))
```

```python
# rotate an array to the right by k steps
class soluition:
    def rotate_arr(self, nums,k):
        k = k % len(nums)
        nums[:] = nums[-k:] + nums[:-k]
        return nums
s7 = soluition()
print(s7.rotate_arr([1,2,3,4,5,6,7], 3))
```
